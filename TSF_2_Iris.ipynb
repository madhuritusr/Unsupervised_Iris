```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "./iris.csv"
df = pd.read_csv(url)
df
```

```python
X = df.iloc[:,1:5].values
X
```

```python
# The elbow method to find clusters
from sklearn.cluster import KMeans
np.random.seed(32)
out = []
for i in range(1,11):
    cluster = KMeans(n_clusters = i, init = 'k-means++', n_init="auto")
    cluster.fit(X)
    out.append(cluster.inertia_)
plt.plot(range(1, 11), out)
plt.title("Elbow Method")
plt.xlabel("No of clusters")
plt.ylabel("out")
plt.xticks()
plt.yticks();
```

The optimum clusters is where the elbow occurs. This is when the within cluster sum of squares (out) doesn't decrease significantly with every iteration which is at **3**

```python
np.random.seed(32)
cluster = KMeans(n_clusters = 3)
cluster.fit(X)
```

```python
y_preds = cluster.predict(X)
y_preds
```

```python
y_preds1 = cluster.fit_predict(X)
y_preds1
```

* fit(): Purpose: This method computes the centroids of the clusters using the given data.
* predict(): Purpose: This method assigns cluster labels to the given data points based on the centroids computed by fit().

* fit_predict(): Purpose: This method performs both fit() and predict() in one step. It computes the centroids and immediately assigns each data point to a cluster.


```python
# Visualizing the first two columns
plt.scatter(X[y_preds == 0,0], X[y_preds == 0,1], # the 0,1 says that we need the cluster 1 from the 2nd column. 
            s = 100, c = 'red', label = 'Iris-setosa') #There are three clusters, thus 0, 1, 2
plt.scatter(X[y_preds == 1, 0], X[y_preds == 1, 1],
           s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(X[y_preds == 2, 0], X[y_preds == 2, 1],
           s = 100, c = 'blue', label = 'Iris-Virginica')

# Plotting the centroids
plt.scatter(cluster.cluster_centers_[:,0], cluster.cluster_centers_[:,1],
           s = 100, c= 'yellow', label = 'centroids')
plt.legend();
```

```python
# Using fit_predict()
plt.scatter(X[y_preds1 == 0,0], X[y_preds1 == 0,1],
           s = 100, c='red', label = 'Iris-setosa')
plt.scatter(X[y_preds1 == 1,0], X[y_preds == 1, 1],
           s = 100, c='green', label = 'Iris-versicolour')
plt.scatter(X[y_preds1 == 2,0], X[y_preds == 2, 1],
           s = 100, c = 'blue', label = 'Iris-virginica')

plt.scatter(cluster.cluster_centers_[:,0], cluster.cluster_centers_[:,1],
           s = 100, c = 'yellow', label = 'centroids')
plt.legend();
```
